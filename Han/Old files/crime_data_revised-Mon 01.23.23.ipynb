```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("CrimeDate.csv")
df
```

```python
df.info()
```

```python
df['year'] = pd.DatetimeIndex(df['date']).year
df.head()
```

```python
df.info()
```

```python
df['year'] = df['year'].astype(int)
df
```

```python
df.info()
```

```python
df_sorted=df.loc[df['year'].isin([2020,2021,2022])].reset_index()
df_sorted
```

```python
df_sample=df_sorted.sample(n=1000)
df_sample
```

```python
df_sample.to_csv("crime_data_revised2.csv")
```

```python
df_sample.to_json("crime_data.json", orient="records", index=True)
```

```python
df_crime_count = df_sample.groupby(['year','primary_type'])['crime_count'].count().reset_index()
df_crime_count 
```

```python
df_arrest_count = df_sample.groupby(['year','primary_type'])['arrest_count'].count().reset_index()
df_arrest_count
```

```python
##df_arrest_count.plot.bar()
```

```python
df_sample.groupby(['primary_type','year'])['arrest_count'].count().unstack().plot.bar(subplots = True, figsize = (12,8))
```

```python
## step 1, get the data
x1 = df_arrest_count.primary_type
x2 = df_arrest_count.year
y = df_arrest_count.arrest_count


# step 2, init the plot
fig = plt.figure(figsize=(10,8))
plt.bar(x1,x2, y, color="coral")

# step 3, customization
plt.title("Number of Arrests from 2020 to 2022")
plt.ylabel("Number of Arrests")
plt.grid(axis="y", alpha=0.25)

# Sets the x limits of the current chart
plt.xticks(rotation=90, fontsize = 10)

# Sets the y limits of the current chart
plt.ylim(0, max(y)+20)

# step 4, saving/showing
plt.show()
```
